```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./breast-cancer.csv')
```

```python
df.drop(columns=['id'],inplace=True)
```

```python
X = df.iloc[:,1:]
y = df["diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
import torch


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
X_train_tensor = torch.from_numpy(X_train).float().to(device)
X_test_tensor = torch.from_numpy(X_test).float().to(device)
y_train_tensor = torch.from_numpy(y_train).float().to(device)
y_test_tensor = torch.from_numpy(y_test).float().to(device)
```

```python
class MySimpleNN():
    def __init__(self,X,device):
        
        self.device = device
        self.weights = torch.rand(X.shape[1],1,dtype=torch.float32,requires_grad=True,device=device)
        self.bias = torch.zeros(1,dtype=torch.float32,requires_grad=True,device=device)
    def forward(self, X):
        z = torch.matmul(X,self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred
    def loss_function(self,y_pred,y_train_tensor):
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred,epsilon,1-epsilon)
        
        loss = -(y_train_tensor * torch.log(y_pred) + (1-y_train_tensor) * torch.log(1-y_pred)).mean()
        return loss
```

```python
learning_rate = 0.1
epochs = 100
```

```python
model = MySimpleNN(X_train_tensor,device)

#forward pass
for epoch in range(epochs):
    y_pred = model.forward(X_train_tensor)
    # print(y_pred)
    
    #loss
    loss = model.loss_function(y_pred,y_train_tensor)
    
    
    #back pass
    loss.backward()
    
    #param update
    with torch.no_grad():
        model.weights-=learning_rate*model.weights.grad
        model.bias-=learning_rate*model.bias.grad
    
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    
    print(f'Epoch: {epoch+1}, Loss: {loss.item()}')
```

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred>0.99).float()
    accuracy = (y_pred==y_test_tensor).float().mean()
    print(accuracy)
```
